## Part 2 - Using an LSTM Network

The first thing we have to do is establish a baseline. To do this, we will run a Long Short-Term Memory neural network on the prices data because LSTM networks have historically worked well in time-series forecasting. We will gather a measure of error from this, and if our model that uses more predictors can beat the error, then we will use the latter model.

What about the fact that the LSTM is training on far more data than the multivariate model? To answer this, we will run our multivariate model sans the consumption data in order to have two models that have been run on the same timeframe. If this model has an error that is close to that of the LSTM, we will likely go with the multivariate model, since it's far easier to interpret, and we can add consumption data later on to hopefully reduce the error further.

Let's start working on the LSTM. We have to import a function that will turn our prices into values that fall within a range of (0,1) since that's what the LSTM network interprets.

```python
from sklearn.preprocessing import MinMaxScaler

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

Now, we have to split the data into a training set and test set. For the Energy Analytics course, we have been tasked with training up until the end of December 2018 and testing on 2019 data. To do that, let's find which index corresponds with the last trading day in December.

```python
prices['1986-01-01':'2019-01-01'].shape
```

Looks like we'll be training the neural network on 8319 data points. Now, we have to split the data into lists of prices that include the previous 60 trading days. This is because the model will use the 60 previous days to get trained, then use the 61st trading day as a response. Let's scale the prices then split the data.

```python
scaler = MinMaxScaler(feature_range = (0, 1))
prices_scaled = scaler.fit_transform(prices)
```

```python
prices = np.array(prices)
prices_training = []
prices_response = []
for i in range(60, 8319):
    prices_training.append(prices_scaled[i-60:i, 0])
    prices_response.append(prices_scaled[i, 0])
```

We have to convert our new lists into numpy arrays, then turn the training set into a 3-dimensional array so that the LSTM network can interpret it (in this case, the third dimension will be 1 because there is only one feature being used for training).

```python
prices_training, prices_response = np.array(prices_training), np.array(prices_response)
prices_training = np.reshape(prices_training, (prices_training.shape[0], prices_training.shape[1], 1))
```

For the LSTM network, we will add 4 layers with 50 nodes each, and add dropout regularization to each to prevent overfitting. Then, we'll finish up with a dense layer that converts the values of the final 50 nodes into a single value that can be used for price prediction.

```python
model = Sequential()
model.add(LSTM(50, return_sequences = True, input_shape = (prices_training.shape[1], 1)))
model.add(Dropout(.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units = 1))
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(prices_training, prices_response, epochs = 10, batch_size = 32)
```

After running the model through 10 epochs, we can predict the held out data (namely indices 8319 and beyond from the original prices dataset), and then compare them with the actual data to calculate an error. To do this, we have to execute the same idea on the test set that we did on the training set by splitting up the data into the 60 previous price points and labelling them with the 61st price point, which will be predicted by the model.

```python
prices_test = []
for i in range(8319, 8476):
    prices_test.append(prices_scaled[i-60:i, 0])

prices_test = np.array(prices_test)
prices_test = np.reshape(prices_test, (prices_test.shape[0], prices_test.shape[1], 1))
```

```python
predictions = model.predict(prices_test)
predictions = scaler.inverse_transform(predictions)
```

Now that we have our predictions, let's plot them on top of the real data to get a visual understanding of how well the neural network performed.

```python
plt.figure(figsize = (13,9))
plt.plot(prices[8319:8476], color = 'blue', label = 'Actual WTI Prices')
plt.plot(predictions, color = 'red', label = 'Predicted WTI Prices')
plt.title('WTI Crude Oil Price Prediction')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
plt.show()
```

Looks great! Now that we have a visual representation, we can calculate the mean squared error to get an understanding of how close the the predictions were to reality.

```python
residuals = (predictions - prices[8319:8474])
mse = round(np.mean(residuals ** 2), 2)
mse
```

The above model looks good, but there's probably something that's been bothering you if you've paid attention to HOW we've actually predicted the data: this prediction isn't actually useful in the real world because it relies on the previous 60 prices to make predictions, and in reality, we don't have access to these datapoints in long-term forecasting. So, if we're going to use a LSTM network, we have to build one that outputs an n x 1 vector that can predict a series of days worth of data. In this case, we're hoping the model can capture patterns over weeks, months, years, and even decades and extrapolate that to rest of 2019 and all of 2020.

After doing some research, there are some methods that can predict multistep time series that involve convolutional neural networks that work in conjunction with the LSTM, but for now, let's stick with using an LSTM network. The idea here is that we will train the model with 152 days of training data for each iteration (rather than 60), and then run a loop of predictions that cuts out the first training point and inserts the most recent prediction. This loop will be run 152 times, and the last prediction will be based completely on prior predicted data rather than data that we supposedly are not supposed to know. This cuts out the need for using test data for predictions, and will show us if the neural network is actually able to predict future data.

```python
#Repeating the same process 

prices_training2 = []
prices_response2 = []
for i in range(157, 8319):
    prices_training2.append(prices_scaled[i-157:i, 0])
    prices_response2.append(prices_scaled[i, 0])
```

```python
prices_training2, prices_response2 = np.array(prices_training2), np.array(prices_response2)
prices_training2 = np.reshape(prices_training, (prices_training.shape[0], 1, prices_training.shape[1]))
```

Let's create a validation set for the model to work with.

```python
prices_xval = []
prices_yval = []
for i in range(8317, 8474):
    prices_xval.append(prices_scaled[i-60:i, 0])
    prices_yval.append(prices_scaled[i, 0])

prices_xval, prices_yval = np.array(prices_xval), np.array(prices_yval)
prices_xval = np.reshape(prices_xval, (prices_xval.shape[0], 1, prices_xval.shape[1]))
```

```python
#New model with one hidden layer and 100 nodes. For this model, the features are the 60 previous price points,
#rather than having one feature with 60 steps.

model2 = Sequential()
model2.add(LSTM(100, input_shape = (prices_training2.shape[1], prices_training2.shape[2])))
model2.add(Dropout(.2))
model2.add(Dense(units = 1))
model2.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model2.fit(prices_training2, prices_response, epochs = 10, batch_size = 32, validation_data = (prices_xval, prices_yval))
```

Here, we're predicting the first "unknown" data point, appending it to the end of the new set of features, and deleting the oldest until we have a set of data that is filled completely with new, predicted data.

```python
prices_test3 = [prices_scaled[8260:8320]]
predicted_prices = []
predicted_prices = np.array(predicted_prices)
                 
for i in range(0, 157):
    prices_test3 = np.array(prices_test3)
    prices_test3 = np.reshape(prices_test3, (prices_test3.shape[0], 1, prices_test3.shape[1]))
    new_prediction = model2.predict(prices_test3)
    prices_test3 = np.append(prices_test3, new_prediction)
    predicted_prices = np.append(predicted_prices, new_prediction)
    prices_test3 = np.delete(prices_test3, 0)
    prices_test3 = np.reshape(prices_test3, (1, prices_test3.shape[0]))

#Transpose the columns into rows to be plotted, then reverse the scaling.
predicted_prices = np.reshape(predicted_prices, (-1,1))
predicted_prices = scaler.inverse_transform(predicted_prices)
```

Here we're just running a similar simulation to the very first one we did, but with the new model.

```python
prices_test_model2 = []
for i in range(8319, 8476):
    prices_test_model2.append(prices_scaled[i-60:i, 0])

prices_test_model2 = np.array(prices_test_model2)
prices_test_model2 = np.reshape(prices_test_model2, (prices_test_model2.shape[0], 1, prices_test_model2.shape[1]))

predictions2 = model2.predict(prices_test_model2)
predictions2 = scaler.inverse_transform(predictions2)
```

Now that we have all of our predictions, let's plot all of them on one graph and compare them.

```python
plt.figure(figsize = (13,9))
plt.plot(prices[8319:8476], color = 'blue', label = 'Actual WTI Prices')
plt.plot(predicted_prices, color = 'orange', label = 'Predicted WTI Prices for Model 3')
plt.plot(predictions2, color = 'red', label = 'Predicted WTI Prices for Model 2')
plt.plot(predictions, color = 'green', label = 'Predicted WTI Prices for Model 1')
plt.title('WTI Crude Oil Price Prediction')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
plt.show()
```

Here we can see that training the model using different numbers of nodes and layers, as well as changing the length of features or how the features are defined in the first place can change our prediction. Clearly, it's difficult to predict a long series of time steps (such as with price prediction), or else everybody would be rich. However, we can clearly that see that there's some shape to the curve, so the model clearly learned something about how the prices act in some way. 

With future tweaking, it's likely that the prediction could get closer to the truth, but that likely takes months, and even years, to accomplish, and the purpose of this notebook is just to explore methods of prediction.

What we can do, though, is use another form of time series prediction to gather future data for each predictor using Facebook's Prophet time series modeling tool, and inputting imports, exports, and consumption to get future predictions.
